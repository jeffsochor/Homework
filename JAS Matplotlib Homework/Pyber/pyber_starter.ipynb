```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
Cdf = pd.read_csv("data/city_data.csv")
Rdf = pd.read_csv("data/ride_data.csv")
# Combine the data into a single dataset
df = pd.merge(Rdf, Cdf, how ="left", on=["city", "city"])
# Display the data table for preview
df.head()
df.shape
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Urban

# Grab just urban data (filtering)
filter_by_type = df["type"] == "Urban"
urban_data = df[filter_by_type]


# Obtain the x and y coordinates for urban cities
# NOTE: x is Total Number of Rides
# y is Average Fare ($)
# Total Number of Drivers
urban_total_rides = urban_data.groupby(["city"]).count()['ride_id']
urban_average_fare = urban_data.groupby(["city"]).mean()['fare']
urban_driver_count = urban_data.groupby(["city"]).count()['driver_count']
UM = urban_data["driver_count"]*10

# Build the scatter plot
plt.scatter(urban_total_rides, urban_average_fare, marker="o", c="red", s=UM, edgecolors="blue")

# Incorporate the other graph properties
plt.title("Urban Ride Data")

# Create a legend
plt.legend(title="Legend")

# Incorporate a text label regarding circle size


# Save Figure
plt.show()


```

```python
#Suburban
filter_by_type = df["type"] == "Suburban"
Suburban_data = df[filter_by_type]

Suburban_total_rides = Suburban_data.groupby(["city"]).count()['ride_id']
Suburban_average_fare = Suburban_data.groupby(["city"]).mean()['fare']
Suburban_driver_count = Suburban_data.groupby(["city"]).count()['driver_count']
SM = Suburban_data["driver_count"]*10

plt.scatter(Suburban_total_rides, Suburban_average_fare, marker="o", c="blue", s=SM, edgecolors="black")
```

```python
#rural
filter_by_type = df["type"] == "Rural"
Rural_data = df[filter_by_type]

Rural_total_rides = Rural_data.groupby(["city"]).count()['ride_id']
Rural_average_fare = Rural_data.groupby(["city"]).mean()['fare']
Rural_driver_count = Rural_data.groupby(["city"]).count()['driver_count']
RM = Rural_data["driver_count"]*10

plt.scatter(Rural_total_rides, Rural_average_fare, marker="o", c="yellow", s=RM, edgecolors="black")
```

```python
plt.scatter(urban_total_rides, urban_average_fare, marker="o", c="coral", s=UM, edgecolors="black",alpha=.7, label="Urban")
plt.scatter(Suburban_total_rides, Suburban_average_fare, marker="o", c="lightblue", s=SM, edgecolors="black",alpha=.7, label="Suburban")
plt.scatter(Rural_total_rides, Rural_average_fare, marker="o", c="gold", s=RM, edgecolors="black",alpha=.7, label="Rural")

plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
# percent total urban fares => sum of urban fares / total fares
# percent total suburban fares => sum of suburban fares / total fares
# percent total rural fares => sum of urban rural / total fares
percent_data = df.groupby(['type']).sum()['fare']
# Build Pie Chart
labels=["Rural", "Suburban", "Urban"]
colors=["gold", "skyblue", "lightcoral"]
explode=(0, 0, .1)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.pie(percent_data, labels=labels,colors=colors, shadow=True, explode=explode, startangle=150, autopct="%1.1f%%")

# Save Figure
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

total_rides = df.groupby(['type']).count()['ride_id']
# Build Pie Chart
labels=["Rural", "Suburban", "Urban"]
colors=["gold", "lightskyblue", "lightcoral"]
explode=(0, 0, .1)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
plt.pie(total_rides, labels=labels,colors=colors, shadow=True, explode=explode, startangle=150, autopct="%1.1f%%")

# Save Figure
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

TotalDrivers = Cdf.groupby(['type']).sum()['driver_count']/Cdf['driver_count'].sum()
# Build Pie Chart
labels=["Rural", "Suburban", "Urban"]
colors=["gold", "lightblue", "lightcoral"]
explode=(0, 0, .1)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
plt.pie(TotalDrivers, labels=labels,colors=colors, shadow=True, explode=explode, startangle=150, autopct="%1.1f%%")

# Save Figure
plt.show()

```
